# Q.2 Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) and make data frame

```python
# Make a request to the IMDB top rated movies page and retrieve the HTML content
page = requests.get("https://www.imdb.com/chart/top")
```

```python
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify())
```

```python
scraped_movie = soup.find_all("td",class_="titleColumn")
scraped_movie
```

```python
#parse movie names
movies = []
for movie in scraped_movie:
    movie= movie.get_text().replace('\n',"")
    movie= movie.strip("")
    movies.append(movie)
movies
```

```python
scraped_ratings = soup.find_all("td",class_="ratingColumn imdbRating" )
scraped_ratings
```

```python
ratings= []
for rating in scraped_ratings:
    rating= rating.get_text().replace('\n', "")
    ratings.append(rating)
ratings
```

```python
scraped_year= soup.find_all("span",class_="secondaryInfo")
scraped_year
```

```python
years = []
for year in scraped_year:
    year =  year.get_text()
    years.append(year)
years
    
```

```python
df = pd.DataFrame({"movie names":movies,"Released Years":years,"IMdb Ratings":ratings})
df = df.head(50)
```

```python
df
```

# Q.4 Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame

# (a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


```python
page4= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup4= BeautifulSoup(page4.content)

Team=[]
Matches=[]
Points=[]
Rating=[]
Ranking=[]
for i in soup4.find_all('tr')[:11]:
    cells = i.find_all('td')
    if len(cells) == 5:
        Ranking.append(cells[0].text.strip())
        Team.append(cells[1].text.strip().replace('\n'," , "))
        Matches.append(cells[2].text.strip())
        Points.append(cells[3].text.strip())
        Rating.append(cells[4].text.strip())
        
print('-'*50)
print("\t Top 10 ODI Teams in Men's Cricket")
print('-'*50)

df_team = pd.DataFrame({'Ranking':Ranking,'Team':Team,'Matches':Matches,'Points':Points,'Rating':Rating})  
df_team
```

```python
#(6) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
#    (i) Headline
#    (ii) Time
#    (iii) News Link

page7= requests.get('https://www.cnbc.com/world/?region=world')
soup7= BeautifulSoup(page7.content)

headline=[]
for i in soup7.find_all('a', class_="LatestNews-headline"):
    headline.append(i.get('title'))

time=[]
for i in soup7.find_all('span', class_="LatestNews-wrapper"):
    time.append(i.text)

News_Link = []
for i in soup7.findAll('a',class_="LatestNews-headline"):
    News_Link.append(i.get('href'))
    
print('-'*90)
print('\t\t\t\t\t News Details')
print('-'*90)

News=pd.DataFrame({'Headline': headline, 'Time': time,'News Link': News_Link})
News
```

Write a python program to scrape mentioned details from dineout.co.in and make data framei) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
#(8) Write a python program to scrape mentioned details from dineout.co.in :
#    (i) Restaurant name
#    (ii) Cuisine
#    (iii) Location
#    (iv) Ratings
#    (v) Image URL

page9= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup9= BeautifulSoup(page9.content)

restaurant_name= []
for i in soup9.find_all('a', class_="restnt-name ellipsis"):
    restaurant_name.append(i.text)

Cuisine=[]
for i in soup9.find_all('div', class_="detail-info"):
    Cuisine.append((i.text.split()[6])+(' ')+(i.text.split()[7]))


location=[]
for i in soup9.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)

Ratings= []
for i in soup9.find_all('div', class_="restnt-rating rating-4"):
    Ratings.append(i.text)

Image_URL= []
for i in soup9.find_all('img', class_="no-img"):
    Image_URL.append(i['data-src'])


print('-'*127)
print('\t\t\t\t\t\t Restaurant Details')
print('-'*127)

dineout=pd.DataFrame({'Restaurant_Name':restaurant_name,'Cuisine':Cuisine,'Location':location,'Ratings':Ratings,'Image_URL':Image_URL})
dineout
```
